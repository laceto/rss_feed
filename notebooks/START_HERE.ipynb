## Setup check — run the cell below first

````python
import sys
from pathlib import Path
from datetime import date

PROJECT_ROOT = Path("..") if Path("../pipeline").exists() else Path(".")
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from pipeline.constants import SECTOR_SUMMARY_FILE, TOPIC_TRENDS_FILE

BRIEFINGS_DIR = PROJECT_ROOT / "data" / "briefings"
today = date.today().isoformat()

checks = {
    "sector_summary.tsv":   SECTOR_SUMMARY_FILE.exists(),
    "topic_trends.tsv":     TOPIC_TRENDS_FILE.exists(),
    f"briefings/{today}.json": (BRIEFINGS_DIR / f"{today}.json").exists(),
}

print(f"Data availability check ({today}):\n")
for name, ok in checks.items():
    status = "OK" if ok else "MISSING"
    print(f"  [{status:^7}]  {name}")

print()
if not checks["sector_summary.tsv"]:
    print("  sector_summary.tsv missing — run the setup cell in query_sector_guide.ipynb to pull from HuggingFace.")
if not checks["topic_trends.tsv"]:
    print("  topic_trends.tsv missing — CI generates this; or run:  just cluster-topics")
if not checks[f"briefings/{today}.json"]:
    print(f"  No briefing for {today} yet — CI runs at ~13:00 UTC on weekdays.")
````
